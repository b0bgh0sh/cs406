We are given that $f(x) = \frac{1}{2}x^TAx+b^Tx+c$; where $A \in R^{n \times n}$ is PSD $(A \succeq 0)$, $b \in R^n$ and $c \in R$. <br><br>

Firstly, we take the gradient of $f(x)$<br>
$\nabla f(x) = Ax+b^T$; and we will show that $b \in range(A)$. <br>
Now, if we set this to 0, we get $Ax = -b^T$. Let's say we have $\tilde{x}$ that solves the equation, such that $\nabla f(\tilde{x}) = 0$. We already know that A is PSD, $\therefore \nabla^2 f(x) = A$ is also a PSD. From both of the aforementioned facts, we can conclude that $f(x) \geq f(\tilde{x}) > -\infty; \forall x \in R^n$. 

Now, for a more formal proof, let us assume that $b \notin range(A)$. We already know that the main function is bounded, and so there must be a $\tilde{x}$ such that $f(x) \geq f(\tilde{x}) > -\infty; \forall x \in R^n$. And since $\tilde{x}$ is a solution, $\therefore \nabla f(\tilde{x}) = 0$ and $\nabla^2 f(\tilde{x})$ is a PSD. <br>
When we compute the $\nabla f(x) = 0$, we get $Ax = -b^T$. But, we already know that there is $\tilde{x}$ exists, and $\nabla f(\tilde{x}) = 0$. Thus we hit a contradiction to our original assumption for this part, that $b \notin A$, which makes it impossible to happen.

Thus, now we can finally conclude that $f$ is bounded below over $R^n$ iff $b \in range(A)$.

## Question 2:

$A \in R^{m \times n}$ and $B \in R^{m \times n}$

Let $A = \begin{bmatrix} A_{1,1} && A_{1,2} && \dots && A_{1,n}\\ A_{2,1} && A_{2,2} && \dots && A_{2,n}\\ \vdots && \vdots && \dots && \vdots \\ A_{m,1} && A_{m,2} && \dots && A_{m,n} \end{bmatrix}$. Let $B = \begin{bmatrix} B_{1,1} && B_{1,2} && \dots && B_{1,n}\\ B_{2,1} && B_{2,2} && \dots && B_{2,n}\\ \vdots && \vdots && \dots && \vdots \\ B_{m,1} && B_{m,2} && \dots && B_{m,n} \end{bmatrix}$

$\therefore A^T = \begin{bmatrix} A_{1,1} && A_{2,1} && \dots && A_{m,1}\\ A_{1,2} && A_{2,2} && \dots && A_{m,2}\\ \vdots && \vdots && \dots && \vdots \\ A_{1,n} && A_{2,n} && \dots && A_{m,n} \end{bmatrix}$

$(A^TB)_{1,1} = A_{1,1}B_{1,1}+A_{2,1}+B_{2,1}+\dots+A_{m,1}B_{m,1}$<br>
$(A^TB)_{2,2} = A_{1,2}B_{1,2}+A_{2,2}+B_{2,2}+\dots+A_{m,2}B_{m,2}$<br>
$\vdots$ <br>
$(A^TB)_{n,n} = A_{1,n}B_{1,n}+A_{2,n}+B_{2,n}+\dots+A_{m,n}B_{m,n}$<br><br>

$\therefore$ We see that $tr(A^TB) = \sum_{i,j}A_{ij}B_{ij}$.

```julia
m = 100;
n = 50;
A = randn(m,n);
B = randn(m,n);
@time tr(A'*B);

@time function rhs()
    s = 0;
    for i in 1:m
        for j in 1:n
            s += A[i,j]*B[i,j];
        end
    end
    return s;
end;
@printf("LHS = %f \n",tr(A'*B));
@printf("RHS = %f",rhs());
```

Above, I used Random Matrices of size $100 \times 50$ to calculate the time and storage complexities of two sides of the equation. Computationaly, they take around the same time, albeit, RHS is a bit faster and takes less storage space. So, RHS seems to be a superior formula.

### a

We are given that $A$ is a diagonal matrix. Now, let us assume that $A \succeq 0$. Stating what's given along with our assumption, we can write $x^TAx$ as $\lambda_1x_1^2+\lambda_2x_2^2+\dots+\lambda_nx_n^2$ for a vector $\{x_1, x_2, \dots, x_n\}$. A is a diagonal matrix, so the diagonal values are the eigenvalues. <br><br>

Since, we are assuming $A \succeq 0$, $x^TAx \geq 0$ has to be true.<br> $\therefore \lambda_1x_1^2+\lambda_2x_2^2+\dots+\lambda_nx_n^2 \geq 0 => \lambda_i \geq 0, \forall i \in \{1, \dots, n\}$ which confirms that for $A \succeq 0$ iff $A_{ii} \geq 0$. In similar way we can prove for $A \succ 0 $. <br><br>

If we assume that $A \succ 0$, then $x^TAx > 0$ has to be true.<br> $\therefore \lambda_1x_1^2+\lambda_2x_2^2+\dots+\lambda_nx_n^2 > 0 => \lambda_i > 0, \forall i \in \{1, \dots, n\}$ which confirms that for $A \succ 0$ iff $A_{ii} > 0$.
